# DQN

The DeepQNetwork class in this project, inheriting from nn.Module, implements a deep Q-learning model for reinforcement learning tasks. The network is designed with a series of convolutional layers followed by fully connected layers to process input states and output Q-values for each possible action. The convolutional layers extract spatial features from the input, while the fully connected layers integrate these features to estimate the Q-values. The forward method defines the data flow through the network. Additionally, the select_action method implements an epsilon-greedy policy for action selection, balancing exploration and exploitation by choosing random actions with a probability of epsilon or the action with the highest Q-value otherwise.








```python
class DeepQNetwork(nn.Module):
    def __init__(self, input_shape, num_actions):
        super(DeepQNetwork, self).__init__()

        self.input_shape = input_shape
        self.num_actions = num_actions

        # Define the convolutional layers
        self.conv_layers = nn.Sequential(
            nn.Conv2d(input_shape[0], 32, kernel_size=8, stride=4),
            nn.ReLU(),
            nn.Conv2d(32, 64, kernel_size=4, stride=2),
            nn.ReLU(),
            nn.Conv2d(64, 64, kernel_size=3, stride=1),
            nn.ReLU()
        )

        # Define the fully connected layers
        self.fc_layers = nn.Sequential(
            nn.Linear(7 * 7 * 64, 512),
            nn.ReLU(),
            nn.Linear(512, self.num_actions)
        )

    def forward(self, x):
        # Pass input through convolutional layers
        x = self.conv_layers(x)
        # Flatten the output from conv layers
        x = x.view(x.size(0), -1)
        # Pass the flattened output through fully connected layers
        x = self.fc_layers(x)
        return x

    def select_action(self, state, epsilon):
        """
        Select an action using epsilon-greedy policy
        """
        # With probability epsilon, select a random action
        if random.random() < epsilon:
            action = random.randrange(self.num_actions)
        else:
            # Otherwise, select the action with the highest Q-value
            state_tensor = torch.FloatTensor(np.float32(state)).unsqueeze(0).to(device)
            q_values = self.forward(state_tensor)
            action = q_values.max(1)[1].item()
        return action
```

# Replay Buffer

The ReplayBuffer class is designed to manage and store transitions encountered during the training of a reinforcement learning agent. It initializes with a specified capacity, using a deque to efficiently handle the buffer's maximum length. The push method stores transitions, including the current state, action, reward, next state, and done flag, with an added dimension for batch processing. The sample method enables random sampling of a batch of transitions from the buffer, crucial for breaking correlation between consecutive experiences and stabilizing training. It returns the sampled transitions as separate arrays for states, actions, rewards, next states, and done flags. The __len__ method returns the current size of the buffer, facilitating buffer management and capacity checks.

```python
class ReplayBuffer(object):
    def __init__(self, capacity):
        """Initialize the replay buffer with a given capacity."""
        # Create a deque with a maximum length to store the buffer
        self.buffer = deque(maxlen=capacity)

    def push(self, state, action, reward, next_state, done):
        """Store a transition in the replay buffer.

        Parameters:
        state: The current state.
        action: The action taken.
        reward: The reward received.
        next_state: The next state after the action.
        done: Whether the episode is done.
        """
        # Add an extra dimension to state and next_state for batch processing
        state = np.expand_dims(state, axis=0)
        next_state = np.expand_dims(next_state, axis=0)

        # Append the transition tuple to the buffer
        self.buffer.append((state, action, reward, next_state, done))

    def sample(self, batch_size):
        """Sample a batch of transitions from the replay buffer.

        Parameters:
        batch_size: The number of transitions to sample.

        Returns:
        A tuple of (states, actions, rewards, next_states, dones).
        """
        # Randomly sample a batch of transitions from the buffer
        transitions = random.sample(self.buffer, batch_size)
        # Unzip the sampled transitions
        state, action, reward, next_state, done = zip(*transitions)

        # Concatenate the state and next_state arrays along the batch dimension
        return np.concatenate(state), action, reward, np.concatenate(next_state), done

    def __len__(self):
        """Return the current size of the replay buffer."""
        return len(self.buffer)
```

# Hyperparameters


This section outlines the configuration parameters for training a deep Q-learning agent. The number of training episodes is set to 500 (NUM_EPISODES). The batch size for sampling from the replay buffer is increased to 64 (BATCH_SIZE), enhancing the learning stability. The discount factor for computing future rewards is set at 0.95 (DISCOUNT_FACTOR), balancing the weighting of immediate and future rewards. The epsilon-greedy exploration strategy starts with an initial epsilon value of 1.0 (INITIAL_EPSILON) and decays to a minimum value of 0.1 (MINIMUM_EPSILON) over 25,000 steps (EPSILON_DECAY_RATE). Training begins only after the replay buffer has accumulated a minimum of 5,000 transitions (MIN_REPLAY_SIZE), with a total buffer capacity set to 50,000 transitions (REPLAY_BUFFER_CAPACITY).

```python
# Configuration Parameters

# Adjusted number of training episodes
NUM_EPISODES = 500

# Increased batch size for sampling from the replay buffer
BATCH_SIZE = 64

# Discount factor for computing future rewards
DISCOUNT_FACTOR = 0.95

# Initial epsilon value for the epsilon-greedy exploration strategy
INITIAL_EPSILON = 1.0

# Minimum epsilon value for the epsilon-greedy strategy
MINIMUM_EPSILON = 0.1

# Epsilon decay factor for the epsilon-greedy strategy
EPSILON_DECAY_RATE = 25000

# Minimum required transitions in the replay buffer before training begins
MIN_REPLAY_SIZE = 5000

# Total capacity of the replay buffer
REPLAY_BUFFER_CAPACITY = 10 * MIN_REPLAY_SIZE
```
